```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Define mean and standard deviation
mu = 0  # Mean
sigma = 1  # Standard deviation

# Create values for the x-axis
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)

# Calculate the PDF for the normal distribution
y = norm.pdf(x, mu, sigma)
```

```python
# Calculate probabilities
print(f"P(X <= 0): {norm.cdf(0, mu, sigma):.4f}") # probability of X <= 0
print(f"P(X <= 1): {norm.cdf(1, mu, sigma):.4f}") # probability of X <= 1
print(f"P(X >= 1): {1-norm.cdf(1, mu, sigma):.4f}") # probability of X >= 1
print(f"P(X >= -1 and X <=1) : {norm.cdf(1, mu, sigma)-norm.cdf(-1, mu, sigma):.4f}") # probability of -1 <= X <= 1
print(f"P(X >= -2 and X <=2) : {norm.cdf(2, mu, sigma)-norm.cdf(-2, mu, sigma):.4f}") # probability of -2 <= X <= 2
print(f"P(X >= -3 and X <=3) : {norm.cdf(3, mu, sigma)-norm.cdf(-3, mu, sigma):.4f}") # probability of -3 <= X <= 3

# Calculate z-scores
x_val = 1.5
z_score = (x_val - mu)/sigma
print(f"The z-score for x = {x_val} is: {z_score}")

# Find the value at the 95th percentile
percentile = norm.ppf(0.95, mu, sigma)
print(f"The value at 95th percentile: {percentile}")
```

```python

# Plot the normal distribution curve
plt.plot(x, y, label='Normal Distribution')
plt.title(f'Normal Distribution (μ={mu}, σ={sigma})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data for PMF (Discrete Case)
dice_values = np.arange(1, 7)
pmf_values = [1/6] * 6

# Data for PDF (Continuous Case)
x_pdf = np.linspace(-4, 4, 500)
pdf_values = norm.pdf(x_pdf, loc=0, scale=1)

# Data for CDF
x_cdf = np.linspace(-4, 4, 500)
cdf_values = norm.cdf(x_cdf, loc=0, scale=1)

# Plot setup
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# PMF Plot
axes[0].bar(dice_values, pmf_values, color='skyblue', alpha=0.8, edgecolor='black')
axes[0].set_title("PMF: Probability Mass Function (6-sided Dice)", fontsize=14)
axes[0].set_xlabel("Dice Value", fontsize=12)
axes[0].set_ylabel("Probability", fontsize=12)
axes[0].set_xticks(dice_values)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# PDF Plot
axes[1].plot(x_pdf, pdf_values, color='orange', label="PDF", linewidth=2)
axes[1].fill_between(x_pdf, pdf_values, alpha=0.3, color='orange')
axes[1].set_title("PDF: Probability Density Function (Normal Distribution)", fontsize=14)
axes[1].set_xlabel("x", fontsize=12)
axes[1].set_ylabel("Density", fontsize=12)
axes[1].legend(fontsize=10)
axes[1].grid(linestyle='--', alpha=0.7)

# CDF Plot
axes[2].plot(x_cdf, cdf_values, color='green', label="CDF", linewidth=2)
axes[2].fill_between(x_cdf, cdf_values, alpha=0.3, color='green')
axes[2].set_title("CDF: Cumulative Distribution Function (Normal Distribution)", fontsize=14)
axes[2].set_xlabel("x", fontsize=12)
axes[2].set_ylabel("Cumulative Probability", fontsize=12)
axes[2].legend(fontsize=10)
axes[2].grid(linestyle='--', alpha=0.7)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

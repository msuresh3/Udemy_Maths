# **Binomial Distribution**
Coin Toss: Tossing a coin 10 times (n=10), where p=0.5. The binomial distribution models the number of heads (successes).

Let‚Äôs consider ùëõ=10 trials with p=0.5.The graph will show the probabilities of observing ùëò successes (ùëò ranging from 0 to ùëõ).

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parameters for the Binomial Distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Possible outcomes (0 to n successes)
k = np.arange(0, n + 1)

# PMF for the binomial distribution
pmf_values = binom.pmf(k, n, p)
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.bar(k, pmf_values, color='skyblue', alpha=0.8, edgecolor='black')
plt.title("Binomial Distribution (n = 10, p = 0.5)", fontsize=14)
plt.xlabel("Number of Successes (k)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(k)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Re-import necessary libraries due to code environment reset
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python

# Example: Average rate of 5 events per unit of time (λ=5)
lambda_value = 5

# Calculate the probability of seeing 0, 1, 2... events up to 15
k_values = np.arange(0, 16)
probabilities = poisson.pmf(k_values, lambda_value)

# Display results in table format
print("k  \t P(X=k)")
for k, p in zip(k_values, probabilities):
    print(f"{k} \t {p:.4f}")

# Create a plot to visualize the distribution
plt.bar(k_values, probabilities)
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability (P(X=k))')
plt.title(f'Poisson Distribution (λ = {lambda_value})')
plt.show()

# Calculate specific probabilities
print(f"\nP(X = 3): {poisson.pmf(3, lambda_value):.4f}")
print(f"P(X <= 3): {poisson.cdf(3, lambda_value):.4f}")
print(f"P(X >= 3): {1 - poisson.cdf(2, lambda_value):.4f}")
```

```python


# Parameters for the Poisson Distribution
lambda_val = 4  # Average rate of occurrence
k_values = np.arange(0, 16)  # Possible number of events (0 to 15)

# PMF for the Poisson distribution
poisson_pmf = poisson.pmf(k_values, lambda_val)

# Plot
plt.figure(figsize=(8, 6))
plt.bar(k_values, poisson_pmf, color='lightcoral', alpha=0.8, edgecolor='black')
plt.title("Poisson Distribution (λ = 4)", fontsize=14)
plt.xlabel("Number of Events (k)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
